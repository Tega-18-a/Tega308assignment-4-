### 1. Introduction
This project aims to predict whether a bank client will subscribe to a term deposit based on demographic and campaign data. The dataset used is the *Bank Marketing dataset* from the UCI Machine Learning Repository, which provides a real-world marketing context. I sampled 5,000 records from the original 45,211 to meet the assignment requirement. The research question is: Can we predict subscription likelihood using features like age, job, and call duration? This analysis could help banks optimize their marketing strategies.

### 2. Dataset Summary
- **Source**: UCI Machine Learning Repository (Moro et al., 2014).
- **Records**: 5,010 (sampled from 45,211).
- **Features**: 17 (e.g., age, job, balance, duration, poutcome).
- **Target Variable**: Subscription (binary: 'no' or 'yes').
- **Key Variables**: Duration (call length), job, balance.

Letâ€™s load and inspect the dataset.

```python
import pandas as pd

# Load the dataset
# Update the path below if bank-full.csv is in a different directory
df = pd.read_csv(r'C:\Users\Deborah Okoh\Documents\Tega\CSC\CSC 308\bank+marketing\bank\bank-full.csv', sep=';')

# Sample 5,000 records
df = df.sample(n=5010, random_state=42)

# Inspect the dataset
print('Dataset Shape:', df.shape)
print('\nDataset Info:')
print(df.info())
print('\nFirst 5 Rows:')
print(df.head())
print('\nStatistical Summary:')
print(df.describe())
```

### 3. Data Cleaning and Preprocessing
- **Issues Encountered**: No missing values were found; however, categorical variables needed encoding.
- **Steps Taken**:
  - Encoded categorical features (e.g., job, education) using LabelEncoder.
  - Scaled numeric features (e.g., age, balance) with StandardScaler.
  - Sampled 5,000 records as required.
- **Outcome**: Cleaned dataset with 5,000 records and 16 features.

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Check for missing values
print('Missing Values:')
print(df.isnull().sum())

# Encode categorical variables
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Encode target variable
df['y'] = df['y'].map({'yes': 1, 'no': 0})

# Scale numeric features
numeric_cols = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Verify the cleaned dataset
print('\nCleaned Dataset Info:')
print(df.info())
print('\nFirst 5 Rows After Preprocessing:')
print(df.head())
```

- **Trends**:
  - Longer call durations correlate with higher subscription rates.
  - Certain jobs (e.g., students) are more likely to subscribe.
- **Correlations**:

### 5. Visualizations
The following visualizations provide insights into the data:

### 6. Machine Learning Model
- **Task**: Classification (predict subscription: yes or no).
- **Model**: Random Forest Classifier with 100 trees, chosen for its ability to handle imbalanced data.
- **Preprocessing**: Already encoded categoricals and scaled numerics; split data into 70/30 train/test.
- **Performance Metrics**:
  - Accuracy: 0.906.
  - Confusion Matrix: [1312 true negatives, 23 false positives; 117 false negatives, 48 true positives].
  - Precision/Recall/F1-score: 0.92/0.98/0.95 for the 'yes' class.
- **Interpretation**: The model achieved high accuracy but low recall for subscriptions due to class imbalance. Duration was the top feature.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Define features and target
X = df.drop('y', axis=1)
y = df['y']

# Split the data (70/30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf_model.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print('\nClassification Report:')
print(classification_report(y_test, y_pred))

# Feature importance
feature_importance = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=False)
print('\nFeature Importance:')
print(feature_importance)
```

### 7. Conclusion
- **Findings**: Call duration and job type are strong predictors of subscriptions, with ~90% accuracy.
- **Implications**: Banks can focus marketing efforts on longer calls and target specific jobs like students and retirees.
- **Limitations**: Class imbalance reduced performance for the minority class; the sampled dataset may miss some patterns.
- **Future Work**: Use SMOTE to address class imbalance, test other models like SVM, or use the full dataset.
