This notebook implements a data science project to predict whether a client subscribes to a term deposit (classification task) using a sampled subset of 5,000 records from the Bank Marketing dataset.

## Import and Inspect Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Load and sample dataset
df = pd.read_csv('C:/Users/Achoma/Documents/CSC/308/Bank_marketing/bank.csv', sep=';')
df = df.sample(n=5000, random_state=42)  # Sample 5,000 records

# Inspect data
print('Shape:', df.shape)
print('\nInfo:')
df.info()
print('\nHead:')
display(df.head())
print('\nDescription:')
display(df.describe())
```

## Clean and Preprocess Data

- Check for missing values.
- Encode categorical variables.
- Scale numeric features.
- Drop irrelevant columns if any.

```python
# Check for missing values
print('Missing Values:')
print(df.isnull().sum())

# Encode categorical variables
le = LabelEncoder()
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Scale numeric features
scaler = StandardScaler()
numeric_cols = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print('Cleaned Data Shape:', df.shape)
display(df.head())
```

## Exploratory Data Analysis (EDA)
- Analyze class distribution.
- Examine correlations.
- Group by key features.

```python
# Class distribution
print('Subscription Class Distribution:')
print(df['y'].value_counts(normalize=True))

# Correlation matrix
corr = df.corr()
print('\nCorrelation Matrix:')
display(corr)

# Group by job and subscription
job_subscription = df.groupby('job')['y'].mean()
print('\nMean Subscription by Job:')
display(job_subscription)
```

## Visualizations

Creating five visualizations: 2 with Matplotlib, 2 with Seaborn, 1 additional.

```python
# Visualization 1: Matplotlib Histogram (Age Distribution)
plt.figure(figsize=(8, 6))
plt.hist(df['age'], bins=20, color='lightblue', edgecolor='black')
plt.title('Age Distribution of Clients')
plt.xlabel('Age (Scaled)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
print('Insight: Most clients are middle-aged, with fewer older clients.')

# Visualization 2: Matplotlib Bar Chart (Subscription by Job)
plt.figure(figsize=(10, 6))
job_subscription.plot(kind='bar', color='salmon')
plt.title('Mean Subscription Rate by Job')
plt.xlabel('Job (Encoded)')
plt.ylabel('Proportion Subscribed')
plt.grid(True)
plt.show()
print('Insight: Certain jobs (e.g., students, retirees) have higher subscription rates.')

# Visualization 3: Seaborn Boxplot (Call Duration by Subscription)
plt.figure(figsize=(8, 6))
sns.boxplot(x='y', y='duration', hue='y', data=df, palette='Set3', legend=False)
plt.title('Call Duration by Subscription Status')
plt.xlabel('Subscribed (0: No, 1: Yes)')
plt.ylabel('Call Duration (Scaled)')
plt.show()
print('Insight: Longer call durations are associated with subscriptions.')

# Visualization 4: Seaborn Heatmap (Correlation Matrix)
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation Matrix of Features')
plt.show()
print('Insight: Duration and poutcome have stronger correlations with subscription.')

# Visualization 5: Seaborn Scatterplot (Age vs. Balance)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='balance', hue='y', size='y', data=df, palette='Set1')
plt.title('Age vs. Balance by Subscription Status')
plt.xlabel('Age (Scaled)')
plt.ylabel('Balance (Scaled)')
plt.show()
print('Insight: Higher balance and certain age groups correlate with subscriptions.')
```

## Machine Learning Model

- Task: Classification (predict subscription: yes or no).
- Model: Random Forest Classifier.
- Metrics: Accuracy, Confusion Matrix, Precision, Recall, F1-score.

```python
# Prepare data
X = df.drop('y', axis=1)
y = df['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)

print('Accuracy:', accuracy)
print('\nConfusion Matrix:')
print(cm)
print('\nClassification Report:')
print(report)

# Visualize Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
